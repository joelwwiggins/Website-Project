```python
#Import data from CSV
jobs_import=pd.read_csv('data job posts.csv',encoding = "utf-8")
resume_import=pd.read_csv('UpdatedResumeDataSet - Copy.csv',engine="python", encoding="utf-8")
```

```python
#Remove all non utf-8 characters
resume_import_clean=resume_import.replace('[^a-zA-Z\d\_.]+',' ',regex=True)
jobs_import_clean=jobs_import.replace('[^a-zA-Z\d\_.]+',' ',regex=True)
```

```python
#Grab a few resumes
resume_desc=resume_import_clean['Resume']
resume_desc_shortlist=resume_desc.iloc[:4]
resume_desc_shortlist
```

```python
#Grab the jobs
post=jobs_import_clean['jobpost']
company=jobs_import_clean['Company']
post_shortlist=post[0:100]
post_shortlist
```

```python
#Generate cosign table
cos_list=pd.concat([resume_desc_shortlist,post_shortlist],ignore_index=False)
cos_list
```

```python
#Generate cosine similarity table
cv = CountVectorizer()
count_matrix=cv.fit_transform(cos_list)
cos_sim_table=cosine_similarity(count_matrix)
cos_sim_table
#Convert to dataframe 
cos_sim_tabledf=pd.DataFrame(cos_sim_table)
```
